```python
# 1) Find female those who are not alone with fair value is less than 60000and greater than 20000.
```

```python
da=data.loc[((data["fare"]>20)&(data["fare"]<60))&(data["sex"]=="female")]
da
```

```python
#using count values 
```

```python
da[['sex','alone']].value_counts()
```

```python
# using pivote table.
```

```python
pivote=da.pivot_table(index=["sex"],values=["alone"],aggfunc="value_counts")
pivote
```

```python
# 2) Find the Third class travled peoples and find the count of teenage and Adult_age and old age .
```

```python
age=[]
for a in data.age:
    if a>=0 and a<10:
        age.append("Child")
    elif a>=10 and a<18:
        age.append("teenage")
    elif a>=18 and a<35:
        age.append("adult")
    else:
        age.append("old")
#print(age)
```

```python
data.insert(2,"age_new",age)
```

```python
data.head()
```

```python
Third_class=data.loc[(data["class"]=="Third")]
```

```python
Third_class
```

```python
Third_class["age_new"].value_counts()
```

```python
# 4) Find Alive Man and childs from 1st and second class with fair of above 25000
```

```python
ds=data.loc[(data["alive"]=="yes")&(data["fare"]>25)]
ds
```

```python
pivot=ds.pivot_table(index=["age_new"],values=["class"],aggfunc='value_counts')
pivot
```
